````python
#Helpers

from qiskit import QuantumCircuit, ClassicalRegister, QuantumRegister

def gatifier(circuit):
    """
    Builds a gate compatible circuit unrolling the input into u3,cx

    Args
    --------------------
    circuit: circuit to turn into a gate
    """
    # Build a new compatible circuit
    from qiskit.transpiler import PassManager
    from qiskit.transpiler.passes import Unroller
    pass_ = Unroller(['u3', 'cx'])
    pm = PassManager(pass_)
    gate_circuit = pm.run(circuit)
    # turn the new circuit into a gate
    gate_circuit.to_gate()
    return gate_circuit

def diffuser(qbits, ancilla_amount=0):
    """
    Creates a gate that performs a diffusion algorithm to the given qubits.
    The input is expected as: 9 qubits to diffuse, ancilla_amount qubits

    Args
    --------------------
    qbits: the amount of qbits to diffuse
    ancilla_amount: the amount of ancilla qubits to use by the MCT with mode=basic
    """
    qc = QuantumCircuit(qbits + ancilla_amount, name="diffuser")
    qlist = [i for i in range(0,qbits)]
    ancilla = [i for i in range(qbits, qbits+ancilla_amount)]
    # Apply transformation |s> -> |00..0> (H-gates)
    qc.h(qlist)
    # Apply transformation |00..0> -> |11..1> (X-gates)
    qc.x(qlist)
    # Do multi-controlled-Z gate
    qc.h(qlist[-1])
    if ancilla_amount > 0:
        qc.mct(qlist[:-1], qlist[-1], ancilla, mode="basic")  # multi-controlled-toffoli
    else:
        qc.mct(qlist[:-1], qlist[-1])
    qc.h(qlist[-1])
    # Apply transformation |11..1> -> |00..0>
    qc.x(qlist)
    # Apply transformation |00..0> -> |s>
    qc.h(qlist)
    qc.to_gate()
    return qc
````

````python
def qram(problem_set, ancilla_amount=0):
    # 4 address, 16 board
    address = QuantumRegister(4,name="address")
    board = QuantumRegister(16,name="board")
    qc = QuantumCircuit(address, board, name="qRAM")
    if ancilla_amount != 0:
        ancilla = QuantumRegister(ancilla_amount,name="ancilla")
        qc.add_register(ancilla)
    # address 0 - 0000
    qc.x([address[0],address[1],address[2],address[3]])
    for x,y in problem_set[0]:
        if ancilla_amount == 0:
            qc.mct(address, board[int(x)*4 + int(y)])
        else:
            qc.mct(address, board[int(x)*4 + int(y)], ancilla_qubits=ancilla, mode="basic")
    qc.x([address[0],address[1],address[2],address[3]])
    # address 1 - 0001
    qc.x([address[0],address[1],address[2]])
    for x,y in problem_set[1]:
        if ancilla_amount == 0:
            qc.mct(address, board[int(x)*4 + int(y)])
        else:
            qc.mct(address, board[int(x)*4 + int(y)], ancilla_qubits=ancilla, mode="basic")
    qc.x([address[0],address[1],address[2]])
    # address 2 - 0010
    qc.x([address[0],address[1],address[3]])
    for x,y in problem_set[2]:
        if ancilla_amount == 0:
            qc.mct(address, board[int(x)*4 + int(y)])
        else:
            qc.mct(address, board[int(x)*4 + int(y)], ancilla_qubits=ancilla, mode="basic")
    qc.x([address[0],address[1],address[3]])
    # address 3 - 0011
    qc.x([address[0],address[1]])
    for x,y in problem_set[3]:
        if ancilla_amount == 0:
            qc.mct(address, board[int(x)*4 + int(y)])
        else:
            qc.mct(address, board[int(x)*4 + int(y)], ancilla_qubits=ancilla, mode="basic")
    qc.x([address[0],address[1]])
    # address 4 - 0100
    qc.x([address[0],address[2],address[3]])
    for x,y in problem_set[4]:
        if ancilla_amount == 0:
            qc.mct(address, board[int(x)*4 + int(y)])
        else:
            qc.mct(address, board[int(x)*4 + int(y)], ancilla_qubits=ancilla, mode="basic")
    qc.x([address[0],address[2],address[3]])
    # address 5 - 0101
    qc.x([address[0],address[2]])
    for x,y in problem_set[5]:
        if ancilla_amount == 0:
            qc.mct(address, board[int(x)*4 + int(y)])
        else:
            qc.mct(address, board[int(x)*4 + int(y)], ancilla_qubits=ancilla, mode="basic")
    qc.x([address[0],address[2]])
    # address 6 - 0110
    qc.x([address[0],address[3]])
    for x,y in problem_set[6]:
        if ancilla_amount == 0:
            qc.mct(address, board[int(x)*4 + int(y)])
        else:
            qc.mct(address, board[int(x)*4 + int(y)], ancilla_qubits=ancilla, mode="basic")
    qc.x([address[0],address[3]])
    # address 7 - 0111
    qc.x(address[0])
    for x,y in problem_set[7]:
        if ancilla_amount == 0:
            qc.mct(address, board[int(x)*4 + int(y)])
        else:
            qc.mct(address, board[int(x)*4 + int(y)], ancilla_qubits=ancilla, mode="basic")
    qc.x([address[0]])
    # address 8 - 1000
    qc.x([address[1],address[2],address[3]])
    for x,y in problem_set[8]:
        if ancilla_amount == 0:
            qc.mct(address, board[int(x)*4 + int(y)])
        else:
            qc.mct(address, board[int(x)*4 + int(y)], ancilla_qubits=ancilla, mode="basic")
    qc.x([address[1],address[2],address[3]])
    # address 9 - 1001
    qc.x([address[1],address[2]])
    for x,y in problem_set[9]:
        if ancilla_amount == 0:
            qc.mct(address, board[int(x)*4 + int(y)])
        else:
            qc.mct(address, board[int(x)*4 + int(y)], ancilla_qubits=ancilla, mode="basic")
    qc.x([address[1],address[2]])
    # address 10 - 1010
    qc.x([address[1],address[3]])
    for x,y in problem_set[10]:
        if ancilla_amount == 0:
            qc.mct(address, board[int(x)*4 + int(y)])
        else:
            qc.mct(address, board[int(x)*4 + int(y)], ancilla_qubits=ancilla, mode="basic")
    qc.x([address[1],address[3]])
    # address 11 - 1011
    qc.x(address[1])
    for x,y in problem_set[11]:
        if ancilla_amount == 0:
            qc.mct(address, board[int(x)*4 + int(y)])
        else:
            qc.mct(address, board[int(x)*4 + int(y)], ancilla_qubits=ancilla, mode="basic")
    qc.x(address[1])
    # address 12 - 1100
    qc.x([address[2],address[3]])
    for x,y in problem_set[12]:
        if ancilla_amount == 0:
            qc.mct(address, board[int(x)*4 + int(y)])
        else:
            qc.mct(address, board[int(x)*4 + int(y)], ancilla_qubits=ancilla, mode="basic")
    qc.x([address[2],address[3]])
    # address 13 - 1101
    qc.x(address[2])
    for x,y in problem_set[13]:
        if ancilla_amount == 0:
            qc.mct(address, board[int(x)*4 + int(y)])
        else:
            qc.mct(address, board[int(x)*4 + int(y)], ancilla_qubits=ancilla, mode="basic")
    qc.x(address[2])
    # address 14 - 1110
    qc.x(address[3])
    for x,y in problem_set[14]:
        if ancilla_amount == 0:
            qc.mct(address, board[int(x)*4 + int(y)])
        else:
            qc.mct(address, board[int(x)*4 + int(y)], ancilla_qubits=ancilla, mode="basic")
    qc.x(address[3])
    # address 15 - 1111
    for x,y in problem_set[15]:
        if ancilla_amount == 0:
            qc.mct(address, board[int(x)*4 + int(y)])
        else:
            qc.mct(address, board[int(x)*4 + int(y)], ancilla_qubits=ancilla, mode="basic")
    # return gate
    qc.to_gate()
    return qc

def incrementer(ancilla_amount=0):
    """
    Counts the amount of ones in a 9 qubit register.
    Qubits are expected as: 9 input, 4 output, 2 ancilla.
    The result is in the shape:
        count[3] ... count[0]
          LSB    ...   MSB
    i.e. 1 = '001'
    """
    q = QuantumRegister(1,"input")
    count = QuantumRegister(3,"count")
    qc = QuantumCircuit(q,count,name="adder")
    # add q[0] to the counter
    i = 0
    if ancilla_amount == 0:
        qc.mct([q,count[1],count[2]], count[0])
        qc.ccx(q, count[2], count[1])
        qc.cx(q,count[2])
    else:
        ancilla = QuantumRegister(ancilla_amount, name="ancilla")
        qc.add_register(ancilla)
        qc.mct([q,count[1],count[2]], count[0], ancilla, mode="basic")
        qc.ccx(q, count[2], count[1])
        qc.cx(q, count[2])
    qc.to_gate()
    return qc

def mark_solvable_board(ancilla_amount=0):
    board = QuantumRegister(16, name='board')
    count = QuantumRegister(3, name='count')
    oracle = QuantumRegister(1, name='oracle')
    qc = QuantumCircuit(board,count,oracle,name="marker")
    if ancilla_amount != 0:
        ancilla = QuantumRegister(ancilla_amount, name="ancilla")
        qc.add_register(ancilla)
    # "conditions" contains the boards that should be empty after applying any of the possible 56 operations
    conditions = [[12, 13, 14, 15], [8, 9, 10, 11], [9, 10, 11, 13, 14, 15], [8, 10, 11, 12, 14, 15], [8, 9, 11, 12, 13, 15], [8, 9, 10, 12, 13, 14], [4, 5, 6, 7], [5, 6, 7, 13, 14, 15], [4, 6, 7, 12, 14, 15], [4, 5, 7, 12, 13, 15], [4, 5, 6, 12, 13, 14], [5, 6, 7, 9, 10, 11], [4, 6, 7, 8, 10, 11], [4, 5, 7, 8, 9, 11], [4, 5, 6, 8, 9, 10], [6, 7, 10, 11, 14, 15], [5, 7, 9, 11, 13, 15], [5, 6, 9, 10, 13, 14], [4, 7, 8, 11, 12, 15], [4, 6, 8, 10, 12, 14], [4, 5, 8, 9, 12, 13], [0, 1, 2, 3], [1, 2, 3, 13, 14, 15], [0, 2, 3, 12, 14, 15], [0, 1, 3, 12, 13, 15], [0, 1, 2, 12, 13, 14], [1, 2, 3, 9, 10, 11], [0, 2, 3, 8, 10, 11], [0, 1, 3, 8, 9, 11], [0, 1, 2, 8, 9, 10], [2, 3, 10, 11, 14, 15], [1, 3, 9, 11, 13, 15], [1, 2, 9, 10, 13, 14], [0, 3, 8, 11, 12, 15], [0, 2, 8, 10, 12, 14], [0, 1, 8, 9, 12, 13], [1, 2, 3, 5, 6, 7], [0, 2, 3, 4, 6, 7], [0, 1, 3, 4, 5, 7], [0, 1, 2, 4, 5, 6], [2, 3, 6, 7, 14, 15], [1, 3, 5, 7, 13, 15], [1, 2, 5, 6, 13, 14], [0, 3, 4, 7, 12, 15], [0, 2, 4, 6, 12, 14], [0, 1, 4, 5, 12, 13], [2, 3, 6, 7, 10, 11], [1, 3, 5, 7, 9, 11], [1, 2, 5, 6, 9, 10], [0, 3, 4, 7, 8, 11], [0, 2, 4, 6, 8, 10], [0, 1, 4, 5, 8, 9], [11, 3, 15, 7], [2, 10, 6, 14], [1, 13, 5, 9], [0, 8, 4, 12]]
    # check if each operation clears the board
    for c in conditions:
        # check that the tiles that meet this condition are empty
        qc.x(c)
        if ancilla_amount == 0:
            qc.mct(c, oracle)
            qc.append(incrementer(), oracle[:] + count[:])
            qc.mct(c, oracle)
        else:
            qc.mct(c, oracle, ancilla, mode="basic")
            qc.append(incrementer(ancilla_amount), oracle[:] + count[:] + ancilla[:])
            qc.mct(c, oracle, ancilla, mode="basic")
        qc.x(c)
    return qc
````

````python
def week3_ans_func(problem_set, reverse=False):
    ##### build your quantum circuit here
    ##### In addition, please make it a function that can solve the problem even with different inputs (problem_set). We do validation with different inputs. 
    address = QuantumRegister(4, name='address')
    board = QuantumRegister(16, name='tile')
    count = QuantumRegister(3, name='count')
    oracle = QuantumRegister(1, name='oracle')
    ancilla = QuantumRegister(4, name='ancilla')
    bits = ClassicalRegister(4,name='bits')
    qc = QuantumCircuit(address, board,count, oracle,ancilla, bits)
    ##init qubits
    qc.h(address)
    # Init QRAM
    qc.append(qram(problem_set,ancilla_amount=len(ancilla)), address[:]+board[:]+ancilla[:])
    
    # store in "count" the amount of solutions of each board
    qc.append(mark_solvable_board(ancilla_amount=len(ancilla)), board[:] + count[:] + [oracle] + ancilla[:])
    
    # flip the phase of the board where "count"=0
    qc.x(oracle)
    qc.h(oracle)
    qc.x(count)
    qc.mct(count, oracle[0], ancilla, mode="basic")
    qc.x(count)
    qc.h(oracle)
    qc.x(oracle)
    
    # uncompute counter of solutions
    qc.append(mark_solvable_board(ancilla_amount=len(ancilla)).inverse(), board[:] + count[:] + [oracle] + ancilla[:])
    
    # uncompute qram
    qc.append(qram(problem_set,ancilla_amount=len(ancilla)).inverse(), address[:]+board[:]+ancilla[:])
    
    # diffuse
    qc.append(diffuser(4, ancilla_amount=4), address[:] + ancilla[:])
    
    # measure the address
#     qc.measure(address[:] + board[:] + count[:] + oracle[:], bits)
    qc.measure(address[:], bits)
    
    # reverse
    qc = qc.reverse_bits() if reverse else qc
    return qc
````
